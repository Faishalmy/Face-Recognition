```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread(r"data/raw/Kate Winslet/017_fb53f84d.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
from mtcnn import MTCNN
from mtcnn.utils.plotting import plot

detector = MTCNN()
faces = detector.detect_faces(img)
plt.imshow(plot(img, faces))
plt.show()
```

```python
x,y,w,h = faces[0]['box']
faces_img = img[y:y+h, x:x+w]
faces_img = cv.resize(faces_img, (160, 160))
plt.imshow(faces_img)
```
